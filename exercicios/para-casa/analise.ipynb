```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_lesb = pd.read_csv("vivlesb.csv", delimiter=';')
```

```python
# Escolaridade por Identificação racial
escolaridade_por_raca = df_lesb.groupby(['Escolaridade', 'Identificação racial']).size().unstack(fill_value=0)
colors = ['#F5F5DC', '#99FF99', '#00CED1', '#008B8B']
fig, ax = plt.subplots(figsize=(15, 10))
escolaridade_por_raca.plot(kind='bar', stacked=True, ax=ax, color=colors)
ax.set_title('Escolaridade por Identificação racial')
ax.set_xlabel('Raça')
ax.legend(title='Raça')
plt.tight_layout()
plt.show()
```

```python
# Renda por região
renda_por_local = df_lesb.groupby(['Local de Residência', 'Faixa de renda mensal pessoal (per capta)']).size().unstack(fill_value=0)
colors = ['#FFE4E1', '#DDA0DD', '#DB7093', '#C71585']
fig, ax = plt.subplots(figsize=(15, 10))
renda_por_local.plot(kind='bar', stacked=True, ax=ax, color=colors)
ax.set_title('Renda Pessoal por Local de Residência')
ax.set_xlabel('Local de Residência')
ax.legend(title='Faixa de renda')
plt.tight_layout()
plt.show()
```

```python
# Tipo de moradia por idade
moradia_por_idade = df_lesb.groupby(['Tipo de moradia', 'Idade']).size().unstack(fill_value=0)
colors = ['#B0E0E6', '#87CEFA', '#87CEEB', '#6495ED', '#1E90FF']
fig, ax = plt.subplots(figsize=(15, 10))
moradia_por_idade.plot(kind='bar', stacked=True, ax=ax, color=colors)
ax.set_title('Moradia por idade')
ax.set_xlabel('Tipo de moradia')
ax.legend(title='Tipo de moradia por idade')
plt.tight_layout()
plt.show()
```
