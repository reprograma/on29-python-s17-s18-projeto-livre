```python
import matplotlib.pyplot as plt
import seaborn as sns

grafico = sns.countplot(x='Sexo', hue='Presença Matemática', data=df_enem_sociais, palette='viridis')
plt.title('Presença no 1º dia em relação ao Sexo')

for i in grafico.patches:
    height = i.get_height()
    percentage = '{:.1f}%'.format(100 * height / len(df_enem_sociais))
    grafico.text(i.get_x() + i.get_width() / 2., height + 0.1, percentage, ha='center')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
grafico = sns.countplot(data=df_enem_sociais, x='Cor/raça', palette='tab10')


for i in grafico.patches:
    height = i.get_height()
    percentage = (height / len(df_enem_sociais)) * 100
    grafico.annotate(f'{percentage:.1f}%', 
                (i.get_x() + i.get_width() / 2., height), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points', 
                fontsize=10, color='black')

plt.title('Distribuição por Cor/Raça')
plt.xlabel('Cor/Raça')
plt.ylabel('Número de Alunos')

plt.show()
```
