Novo Dataframe com os top 10 por streams

```python
df['streams'] = df['streams'].astype(str)

# Limpar a coluna 'streams'
df['streams'] = df['streams'].str.replace(r'[^\d]', '', regex=True)  # Remoção caracteres não numéricos
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')  # Converção para numérico, definindo erros como NaN
df = df.dropna(subset=['streams'])  # Remove linhas onde 'streams' não pôde ser convertido
df['streams'] = df['streams'].astype(float)  # Converte a coluna para float


# Agrupar por 'track_name' e a coluna correta do nome do artista e somar os 'streams'
top_10 = df.groupby(['track_name', 'artist(s)_name'])['streams'].sum().sort_values(ascending=False).head(10).reset_index()
top_10 = top_10.reset_index(drop=True)
top_10.index = top_10.index + 1
top_10
```

```python

df_streams = df['streams']
df_sorted = df_streams.sort_values(ascending=False)
df_sorted
```

Verificação de relação com o tom entre as mais tocadas 

```python
# Passo 1: Converter a coluna 'key' para um formato categórico
df['key'] = df['key'].astype('category')

# Agrupar por tom e calcular a média de streams
media_streams_por_tom = df.groupby('key')['streams'].mean().sort_values(ascending=False)

# Passo 3: Visualizar os dados
plt.figure(figsize=(10, 6))
media_streams_por_tom.plot(kind='bar', color='skyblue')
plt.xlabel('Tom')
plt.ylabel('Média de Streams')
plt.title('Média de Streams por Tom das Músicas')
plt.xticks(rotation=45)
plt.show()
```

Dataframe das top 10 com tons

```python
tons_df = df[['track_name', 'artist(s)_name', 'key']].drop_duplicates()
top_10_with_keys = pd.merge(top_10, tons_df, on=['track_name', 'artist(s)_name'])
top_10_with_keys = top_10_with_keys.reset_index(drop=True)
top_10_with_keys.index = top_10_with_keys.index + 1
top_10_with_keys
```

Foi verificado que o tom A, o mais utilizado na média de streams, correspode a 57,89% das top 10.

```python
# Encontrar o total de streams das 10 músicas mais tocadas
total_streams_top_10 = top_10_with_keys['streams'].sum()

# Agrupar por 'key' e somar os streams
streams_per_key = top_10_with_keys.groupby('key')['streams'].sum().reset_index()

# Calcular o percentual de cada tom
streams_per_key['percentual'] = (streams_per_key['streams'] / total_streams_top_10) * 100

# Passo 4: Redefinir o índice para começar do 1
streams_per_key_reset = streams_per_key.reset_index(drop=True)
streams_per_key_reset.index = streams_per_key_reset.index + 1
streams_per_key_top_10 = streams_per_key_reset.head(10)
streams_per_key_top_10
```

Analise da dançabilidade

```python
# Converter a coluna 'danceability_%' para um formato categórico
df['danceability_%'] = df['danceability_%'].astype('category')

# Agrupar por dançabilidade e calcular a média de streams
media_streams_por_dance = df.groupby('danceability_%')['streams'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
media_streams_por_dance.plot(kind='bar', color='skyblue')
plt.xlabel('Dançabilidade')
plt.ylabel('Média de Streams')
plt.title('Média de Streams por dançabilidade')
plt.xticks(rotation=45)
plt.show()
```

```python
# Incluindo dançabilidade no dataframe
dance_df = df[['track_name', 'artist(s)_name', 'key', 'danceability_%']].drop_duplicates()
top_10_with_keys_dance = pd.merge(top_10, dance_df, on=['track_name', 'artist(s)_name'])
top_10_with_keys_dance = top_10_with_keys_dance.reset_index(drop=True)
top_10_with_keys_dance.index = top_10_with_keys_dance.index + 1
top_10_with_keys_dance
```

```python
# Garantir que 'danceability_%' seja numérico
df['danceability_%'] = pd.to_numeric(top_10_with_keys_dance['danceability_%'], errors='coerce')

# Obter as 20 maiores dançabilidades
top_20_danceability = df[['danceability_%']].drop_duplicates().nlargest(20, 'danceability_%')

# Filtrar o DataFrame para incluir apenas essas 20 dançabilidades
filtered_df = df[df['danceability_%'].isin(top_20_danceability['danceability_%'])]

# Calcular a média de streams para cada valor de 'danceability_%'
media_streams_por_dance_top_20 = filtered_df.groupby('danceability_%')['streams'].mean().sort_values(ascending=False)


plt.figure(figsize=(12, 8))
media_streams_por_dance_top_20.plot(kind='bar')
plt.xlabel('Dançabilidade (%)')
plt.ylabel('Média de Streams')
plt.title('Média de Streams para as 20 Maior Dançabilidades')
plt.show()
```

Foi detectado relação entre dançabilidade e as mais tocadas.

Identificar e analisar a presença das top 10 músicas nas playlists de diferentes plataformas

```python
# transformando coluna do tipo objeto em inteiro e retirando a virgula
df['in_deezer_playlists'] = df['in_deezer_playlists'].str.replace(',', '').astype(int)
```

```python
# Identificar as top 10 músicas
top_10_musicas = df.groupby(['track_name', 'artist(s)_name'])['streams'].sum().sort_values(ascending=False).head(10).reset_index()

# Filtrar o DataFrame original para apenas as top 10 músicas
df_top_10_musicas = df[df['track_name'].isin(top_10_musicas['track_name'])]

# Verificar presença nas playlists de cada plataforma
df_top_10_musicas['in_spotify_playlists'] = df_top_10_musicas['in_spotify_playlists'] > 0
df_top_10_musicas['in_apple_playlists'] = df_top_10_musicas['in_apple_playlists'] > 0
df_top_10_musicas['in_deezer_playlists'] = df_top_10_musicas['in_deezer_playlists'] > 0

# Exibir o resultado e ordenar por streams em ordem decrescente
df_top_10_musicas_resultado = df_top_10_musicas[['track_name', 'artist(s)_name', 'streams', 'in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists']]
df_top_10_musicas_resultado = df_top_10_musicas_resultado.drop_duplicates(subset=['track_name', 'artist(s)_name'])
df_top_10_musicas_resultado = df_top_10_musicas_resultado.sort_values(by='streams', ascending=False)
df_top_10_musicas_resultado = df_top_10_musicas_resultado.reset_index(drop=True)
df_top_10_musicas_resultado.index = df_top_10_musicas_resultado.index + 1
df_top_10_musicas_resultado

```

Foi verificado que em 100% dos 10 maiores streams, as músicas foram adicionadas as playlists e estão em mais duas outras plataformas.
