```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Abertura das base_estudantes e base_ies

```python
df_estudantes = pd.read_csv('base_estudantes.csv')
df_estudantes
```

```python
df_ies = pd.read_csv('base_ies.csv')
df_ies
```

Somatória de linhas duplicadas

```python
df_ies.duplicated(subset='CO_IES').sum()
```

Limpeza das linhas duplicadas do df_ies

```python
df_ies = df_ies.drop_duplicates(subset='CO_IES')
df_ies
```

Junção dos df_estudantes e df_ies, criando df_completo

```python
df_completo = pd.merge(df_estudantes, df_ies, on=['CO_IES'], how='inner')
df_completo
```

Organização crescente da coluna CO_IES (Código das Isntituições de Ensino)

```python
df_completo = df_completo.sort_values(by= 'CO_IES', ascending= True)
df_completo
```

Filtro (na coluna TP_CATEGORIA_ADMINISTRATIVA) considerando a Lei de Cotas (Lei 12.711/2012): 
1. Pública Federal

```python
df_completo = df_completo.loc[df_completo['TP_CATEGORIA_ADMINISTRATIVA'] == 1]
df_completo
```

Conversão do df para arquivo .csv

```python
df_completo.to_csv('base_completa.csv', index= False)
```

**Análise 1**

Taxa de permanência do total de ingressantes e total de concluintes em 2012, 2017 e 2022

```python
df_status = df_completo.groupby('NU_ANO_CENSO').agg({
    'QT_ING': 'sum',
    'QT_CONC': 'sum'
}).reset_index()
```

```python
df_status['TAXA_PERMANENCIA'] = df_status['QT_CONC'] / df_status['QT_ING']
df_status
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_status['NU_ANO_CENSO'], df_status['TAXA_PERMANENCIA'], marker='o', linestyle='-', color='b')
plt.title('Taxa de Permanência Estudantil (Concluintes / Ingressantes)')
plt.xlabel('Ano')
plt.ylabel('Taxa de Permanência')
plt.ylim(0, 1)
plt.xticks(df_status['NU_ANO_CENSO'])  # Mostrar apenas os anos específicos no eixo x
for i in range(len(df_status)):
    plt.text(df_status['NU_ANO_CENSO'][i], df_status['TAXA_PERMANENCIA'][i] + 0.02, round(df_status['TAXA_PERMANENCIA'][i], 2), ha='center')
plt.grid(True)
plt.show()
```

**Análise 2**

Taxa de permanência considerando gêneros em 2012, 2017 e 2022. Obs.: a base de dados traz informações apenas dos gêneros feminino e masculino. 

```python
df_gen = df_completo.groupby('NU_ANO_CENSO').agg({
    'QT_ING_FEM': 'sum',
    'QT_ING_MASC': 'sum',
    'QT_CONC_FEM': 'sum',
    'QT_CONC_MASC': 'sum'
}).reset_index()

df_gen['TAXA_PERMANENCIA_FEM'] = df_gen['QT_CONC_FEM'] / df_gen['QT_ING_FEM']
df_gen['TAXA_PERMANENCIA_MASC'] = df_gen['QT_CONC_MASC'] / df_gen['QT_ING_MASC']
```

```python
bar_width = 0.35
index = df_gen['NU_ANO_CENSO']

plt.figure(figsize=(12,6))

bar1 = plt.bar(index - bar_width/2, df_gen['TAXA_PERMANENCIA_FEM'], bar_width, color='lightcoral', label='Taxa de Permanência Feminino')
bar2 = plt.bar(index + bar_width/2, df_gen['TAXA_PERMANENCIA_MASC'], bar_width, color='lightblue', label='Taxa de Permanência Masculino')

plt.xlabel('Ano')
plt.ylabel('Taxa de Permanência')
plt.title('Taxa de Permanência Estudantil por Gênero (Concluintes / Ingressantes)')
plt.xticks(index, ['2012', '2017', '2022'])
plt.ylim(0, 1)
plt.legend()

for i in range(len(index)):
    plt.text(index[i] - bar_width/2, df_gen['TAXA_PERMANENCIA_FEM'][i] + 0.02, round(df_gen['TAXA_PERMANENCIA_FEM'][i], 2), ha='center')
    plt.text(index[i] + bar_width/2, df_gen['TAXA_PERMANENCIA_MASC'][i] + 0.02, round(df_gen['TAXA_PERMANENCIA_MASC'][i], 2), ha='center')

plt.show()
```

**Análise 3**

Relação de ingressantes Pretos, Pardos, Indígenas, Brancos, Amarelos em 2012, 2017 e 2022 - questões sobre o incentivo de ingresso no ensino superior para negros, pardos e indígenas

```python
df_raca = df_completo.groupby(['NU_ANO_CENSO']).agg({
    'QT_ING_PRETA': 'sum',
    'QT_ING_PARDA': 'sum',
    'QT_ING_INDIGENA': 'sum',
    'QT_ING_BRANCA': 'sum',
    'QT_ING_AMARELA': 'sum',
    'QT_ING_CORND': 'sum'
}).reset_index()

df_raca
```

```python
df_raca_long = df_raca.melt(id_vars='NU_ANO_CENSO', 
                            value_vars=['QT_ING_PRETA', 'QT_ING_PARDA', 'QT_ING_INDIGENA', 'QT_ING_BRANCA', 'QT_ING_AMARELA', 'QT_ING_CORND'], 
                            var_name='Raça/Cor', 
                            value_name='Quantidade')

df_raca_long['Raça/Cor'] = df_raca_long['Raça/Cor'].replace({
    'QT_ING_PRETA': 'Preta',
    'QT_ING_PARDA': 'Parda',
    'QT_ING_INDIGENA': 'Indígena',
    'QT_ING_BRANCA': 'Branca',
    'QT_ING_AMARELA': 'Amarela',
    'QT_ING_CORND': 'Cor Não Declarada'
})

plt.figure(figsize=(14, 8))
sns.barplot(data=df_raca_long, x='NU_ANO_CENSO', y='Quantidade', hue='Raça/Cor')

plt.xlabel('Ano')
plt.ylabel('Quantidade de Ingressos')
plt.title('Ingressos de Estudantes por Raça/Cor (2012, 2017, 2022)')
plt.legend(title='Raça/Cor')
plt.show()
```
