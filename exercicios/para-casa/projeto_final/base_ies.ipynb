```python
import pandas as pd
```

```python
arquivo = 'MICRODADOS_CADASTRO_IES_2012.CSV'
df_2012 = pd.read_csv(arquivo, encoding = 'latin-1', sep= ';')
df_2012
```

```python
arquivo2 = 'MICRODADOS_CADASTRO_IES_2017.CSV'
df_2017 = pd.read_csv(arquivo2, encoding = 'latin-1', sep= ';')
df_2017
```

```python
arquivo3 = 'MICRODADOS_ED_SUP_IES_2022.CSV'
df_2022 = pd.read_csv(arquivo3, encoding = 'latin-1', sep= ';')
df_2022
```

```python
df2012_final = df_2012[['CO_IES','NO_IES']]
df2012_final
```

```python
df2012_final.isnull().sum()
```

```python
df2017_final = df_2017[['CO_IES','NO_IES']]
df2017_final
```
