```python
quantidade_mulheres_2022 = 1056
quantidade_mulheres_2023 = 1293
quantidade_homens_2022 = 3194
quantidade_homens_2023 = 3975

labels = ["2022", "2023"]
mulheres = [quantidade_mulheres_2022, quantidade_mulheres_2023]
homens = [quantidade_homens_2022, quantidade_homens_2023]

width = 0.35
x = range(len(labels))

plt.bar(x, mulheres, width, label='Mulheres')
plt.bar([p + width for p in x], homens, width, label='Homens')

plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.title('Quantidade de Homens e Mulheres em Dados (2022 X 2023)')
plt.xticks([p + width/2 for p in x], labels)
plt.legend()

plt.show()
```
