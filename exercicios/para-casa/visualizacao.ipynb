```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('base_final_tratada.csv')
df
```

```python
df.describe().T
```

```python
df.columns
```

```python
df['Data_Medicao'] = pd.to_datetime(df['Data_Medicao'])


df['Ano'] = df['Data_Medicao'].dt.year

df_agrupado = df.groupby('Ano')['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 6))
plt.plot(df_agrupado['Ano'], df_agrupado['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'], marker='o', linestyle='-', color='b')
plt.title('Precipitação Total Mensal por Ano')
plt.xlabel('Ano')
plt.ylabel('Precipitação Total (mm)')
plt.grid(True)
plt.xticks(df_agrupado['Ano'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Data_Medicao', y='PRECIPITACAO_TOTAL_MENSAL_AUT_mm_', data=df)
plt.title('Precipitação Total Mensal por Data')
plt.xlabel('Data de Medição')
plt.ylabel('Precipitação Total (mm)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'], bins=20, edgecolor='black')
plt.title('Distribuição da Precipitação Total Mensal')
plt.xlabel('Precipitação Total (mm)')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PRESSAO_ATMOSFERICA_MEDIA_MENSAL_AUT_mB_', y='PRECIPITACAO_TOTAL_MENSAL_AUT_mm_', data=df)
plt.title('Pressão Atmosférica vs Precipitação Total Mensal')
plt.xlabel('Pressão Atmosférica Média (mB)')
plt.ylabel('Precipitação Total (mm)')
plt.tight_layout()
plt.show()
```
