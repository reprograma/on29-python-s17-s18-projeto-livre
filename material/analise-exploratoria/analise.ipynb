```python
import pandas as pd
import squarify
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_indigena = pd.read_csv('terra_indigena.csv')
df_indigena
```

```python
df_indigena.isnull().sum()
```

```python
df_indigena[['id_geografico', 'id_terra_indigena', 'area', 'reestudo', 'indicador_faixa_fronteira', 'unidade_administrativa', 'geometria']] = df_indigena[['id_geografico', 'id_terra_indigena', 'area', 'reestudo', 'indicador_faixa_fronteira', 'unidade_administrativa', 'geometria']].fillna(" ")
df_indigena.isnull().sum()
```

```python
df_indigena.columns
```

Drop das colunas não necessárias

```python
df_final = df_indigena.drop(['reestudo', 'indicador_faixa_fronteira', 'id_unidade_administrativa', 'sigla_unidade_administrativa', 'unidade_administrativa', 'geometria'], axis=1)
df_final
```

Etnias no Brasil

```python
df_etnias_brasil = df_final['etnia'].count()
print(f'Existem {df_etnias_brasil} etnias no Brasil.')
```

Situação dos territórios indígenas no Brasil

```python
areas = df_final.groupby('fase')['terra_indigena'].count()

percentual = df_final['terra_indigena'].value_counts(normalize=True) * 100

plt.title('Situação da Terra Indigena')

# Gráfico de pizza
areas.plot(kind='pie', autopct='%1.1f%%')

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

Situação territorial por estado

```python
agrup_df = df_final.pivot_table(index='sigla_uf', columns='fase', values='terra_indigena', aggfunc='count', fill_value=0)

n = len(agrup_df)
pos = np.arange(n)
largura=0.25

plt.figure(figsize=(12, 8)) # definindo o tamanho do gráfico

for i, fase in enumerate(agrup_df.columns):
    plt.bar(pos + i * largura - largura, agrup_df[fase], largura, label=fase)

plt.xlabel('Sigla UF')
plt.ylabel('Ocorrência')
plt.title('Situação Territorial Indigena por Estado')

plt.xticks(pos, agrup_df.index, rotation=90)
plt.tight_layout()
plt.legend()
```

Etnias por estado

```python
df_etnias_estadual = df_final.groupby('sigla_uf')['etnia'].count().sort_values(ascending=False)
barras = df_etnias_estadual.plot(kind='bar')

# criar a etiqueta das barras
for i in barras.containers:
    barras.bar_label(i, label_type='edge')
```

Distribuição das etnias no estado com maior numero de etnias (Amazonas)

```python
df_amazonas = df_final[df_final['sigla_uf'] == 'AM']
```

```python
etnias_am = df_amazonas.groupby('etnia').size().sort_values(ascending=False).reset_index(name='Quantidade')
etnias_am
```

Distribuição de Territórios Indígenas nos municípios do Amazonas

```python
df_terras_indigenas = df_amazonas.groupby('nome_municipio')['terra_indigena'].count().sort_values(ascending=False)
df_terras_indigenas
```

etnias existentes no municipio com maior numero de territorio indígena

```python
df_st_isabel = df_amazonas[df_amazonas['nome_municipio'] == 'Santa Isabel do Rio Negro']
```

```python
df_st_isabel_etnias = df_st_isabel.groupby('etnia')['etnia'].count().sort_values(ascending=False)

squarify.plot(sizes = df_st_isabel_etnias, label=df_st_isabel_etnias.index, alpha=0.7)
```

Situação dessas etnias

```python
df_fase_st_isabel = df_st_isabel.groupby('fase')['etnia'].count().sort_values(ascending=False)

barras = df_fase_st_isabel.plot(kind='bar', edgecolor='lightblue', color='lightpink')

# criar a etiqueta das barras
for i in barras.containers:
    barras.bar_label(i, label_type='edge')

plt.xticks(rotation=0)
```
