##Junção dos Datasets

```python
df_minas = pd.concat([df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023])
df_minas
```

##Limpeza e Tratamento dos Dados

```python
#Visualizando as informações do dataframe

df_minas.info()
```

```python
#Filtrando somente as colunas que serão relevantes para as análises que desejamos realizar

df_minas = df_minas.drop(['DT_NASC', 'CS_SEXO'], axis=1) #o nascimento é redundante (já temos a idade) e o sexo é feminino em todos os casos
df_minas
```

```python
#Convertendo os valores na coluna 'NU_IDADE_N' (idade da vítima) e 'DT_NOTIFIC' (data da ocorrência) de objetos para inteiros/datetime

df_minas['NU_IDADE_N'] = pd.to_numeric(df_minas['NU_IDADE_N'])
df_minas['DT_NOTIFIC'] = pd.to_datetime(df_minas['DT_NOTIFIC'], dayfirst=True)

df_minas.info() #checando se as alterações foram mesmo realizadas
```

```python
#Checando a quantidade de valores nulos em cada coluna

df_minas.isnull().sum()
```

Como se trata de um assunto delicado, preferimos não dropar nenhum registro por falta de alguma informação, visto que os momentos de denúncia são geralmente muito doloridos para vítimas, que acabam revivendo todo o trauma, e, não por isso, devem ter seus relatos desconsiderados.

```python
#Atribuindo aos campos NaN o valor 'Não informado'

df_minas = df_minas.fillna('Não informado')
df_minas
```

```python
#Rechecando a quantidade de valores nulos no dataframe

df_minas.isnull().sum()
```

###Criando novas colunas

####Mês e ano da ocorrência

```python
#Criando novas colunas indicando mês e ano da ocorrência

df_minas['ANO_NOTIFIC'] = df_minas['DT_NOTIFIC'].dt.year
df_minas['MES_NOTIFIC'] = df_minas['DT_NOTIFIC'].dt.month
df_minas
```

####Participação da vítima na comunidade LGBT

```python
#Vamos verificar quais as possibilidades de sexualidades dentro da coluna 'ORIENT_SEX'

df_minas['ORIENT_SEX'].unique()
```

```python
#Vamos verificar quais as possibilidades de identidade de gênero dentro da coluna 'IDENT_GEN'

df_minas['IDENT_GEN'].unique()
```

```python
#Sabendo das possibilidades, podemos criar uma nova coluna dentro do dataset

df_minas.loc[df_minas['ORIENT_SEX'].isin(['Homossexual', 'Bissexual']), 'LGBT'] = 'Sim'
df_minas.loc[df_minas['IDENT_GEN'].isin(['Transsexual Homem', 'Transsexual Mulher', 'Travesti']), 'LGBT'] = 'Sim'
df_minas['LGBT'] = df_minas['LGBT'].fillna('Não')
df_minas
```

##Exportação da Base

```python
#Resetando os índices do dataframe

df_minas = df_minas.reset_index(drop=True)
df_minas
```

```python
#Criando um arquivo .csv com os dados já juntos, limpos e tratados (para uso no Tableau, por exemplo)

df_minas.to_csv('dataframe_minas.csv', index=False)
```

##Visualizações

###**Visualização 1:** Distribuição das ocorrências ao longo de cada ano

```python
#Contando o número de ocorrência por ano e mês

contagem_ocorrencias = df_minas.groupby(['ANO_NOTIFIC', 'MES_NOTIFIC']).size().unstack()
contagem_ocorrencias
```

```python
#Encontrando o mês com mais ocorrências para cada ano

meses_max_ocorrencias = contagem_ocorrencias.idxmax(axis=1) #encontrando o mês com maior número de ocorrências para cada ano
max_ocorrencias = contagem_ocorrencias.max(axis=1) #encontrando a quantidade de notificações no mês com maior número de ocorrências
```

```python
#Montando a visualização

plt.figure(figsize=(10,8))

for ano in contagem_ocorrencias.index:
  plt.plot(range(1, 13), contagem_ocorrencias.loc[ano], linestyle='-', label=ano) #plotando as linhas
  plt.scatter(meses_max_ocorrencias[ano], max_ocorrencias[ano], color='black', zorder=5) #plotando as bolinhas indicadoras do mês com mais ocorrências

plt.title('Distribuição das ocorrências ao longo de cada ano')
plt.xlabel('Mês')
plt.ylabel('Quantidade de ocorrências')
plt.xticks(range(1,13), labels = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'], rotation=45)
plt.grid(True)
plt.legend(title='Ano', loc='upper right')

plt.show()
```

###**Visualização 2:** Qual o top 5 de cidades em que mais foram registrados casos?

```python
#Contando o número de casos por cidade

casos_cidade = df_minas['ID_MN_RESI'].value_counts()
casos_cidade
```

```python
#Obtendo as cinco cidades com mais casos

top_5 = casos_cidade.head(5)
top_5
```

```python
#Montando a visualização

plt.figure(figsize=(10,8))

sns.barplot(top_5, color='lightblue')

plt.grid(True, axis='y')
plt.title('Top 5 Cidades - Ocorrências')
plt.xlabel('Cidade')
plt.ylabel('Nº de Ocorrências')
plt.xticks(rotation=45)

plt.show()
```

###**Visualização 3:** Quais tipos de violência mais frequentes contra mulheres heterosexuais? E entre mulheres LGBT?

```python
#Substituindo os 'Sim' e 'Não' das colunas de violência por uns e zeros para poder somar posteriormente

colunas_violencia = ['LES_AUTOP', 'VIOL_FISIC', 'VIOL_PSICO', 'VIOL_SEXU']
df_minas[colunas_violencia] = df_minas[colunas_violencia].replace({'Sim': 1, 'Não': 0, 'Não informado': 0, 'Ignorado': 0})
```

```python
#Agrupando as informações de cada categoria (LGBT) e subcategoria (tipo de violência)

df_agrupado = df_minas.groupby('LGBT')[colunas_violencia].sum()
df_agrupado
```

```python
#Montando a visualização

df_agrupado.plot(kind='bar', figsize=(10, 8))

plt.title('Distribuição dos tipos de violência entre mulheres LGBT e não-LGBT')
plt.xlabel('A vítima faz parte da comunidade LGBT?')
plt.ylabel("Nº de Ocorrências")
plt.xticks(rotation=360)
plt.grid(True, axis='y')
plt.legend(title='Tipo de Violência', labels=['Lesão Autoprovocada', 'Violência Física', 'Violência Psicológica', 'Violência Sexual'])

plt.show()
```
