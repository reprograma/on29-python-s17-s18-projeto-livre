Importando as bibliotecas que serão utilizadas no projeto.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Transformando os arquivos xlsx em csv


```python
# Lista de arquivos Excel e seus correspondentes arquivos CSV
excel_files = ['esport_regiao.xlsx', 'esporte_genero.xlsx', 'esporte_idade.xlsx', 'pratica_geral.xlsx', 'frequencia_regiao.xlsx', 'motivo_regiao.xlsx']
csv_files = ['esport_regiao.csv', 'esporte_genero.csv', 'esporte_idade.csv', 'pratica_geral.csv', 'frequencia_regiao.csv', 'motivo_regiao.csv']

# Iterar sobre os arquivos para converter
for excel, csv in zip(excel_files, csv_files):
    # Ler o arquivo Excel
    df_esporte = pd.read_excel(excel)
    # Converter para CSV
    df_esporte.to_csv(csv, index=False)
```

Armazenando o aqrquivo csv em uma variável de DataFrame e visualizando-as 

```python
df_esp_reg = pd.read_csv('esport_regiao.csv')
df_esp_reg.head(5)
```

```python
df_esp_gen = pd.read_csv('esporte_genero.csv')
df_esp_gen.head(5)
```

```python
df_esp_idade = pd.read_csv('esporte_idade.csv')
df_esp_idade.head(5)
```

```python
df_frequencia = pd.read_csv('frequencia_regiao.csv')
df_frequencia.head(7)
```

```python
df_motivo = pd.read_csv('motivo_regiao.csv')
df_motivo.head(7)
```

```python
df_pratica = pd.read_csv('pratica_geral.csv')
df_pratica.head(5)
```

Realizando os Joins para combinação dos DataFrames

```python
#realizando join inner, para combinar os valores correspondentes das duas tabelas
df_esporte = pd.merge(df_esp_gen, df_esp_reg, on=['Principal esporte praticado'], how='inner')
df_esporte
```

```python
df_esporte = pd.merge(df_esporte, df_esp_idade, on=['Principal esporte praticado'], how='inner')
df_esporte
```

```python
#realizando o join preservando a tabela da direito, pois tem uma linha a mais que a da esquerda e preciso de todos os dados
df_final = pd.merge(df_frequencia, df_motivo, on=['id_genero'], how='right')
df_final
```

Convertendo e salvando os DataFrames combinados por join em um arquivo csv


```python
df_esporte.to_csv('esporte.csv', index=False)
df_final.to_csv('frequencia_motivo.csv', index=False)
```

Gerando as visulizações da análise

```python
# Gráfico de quantidade por gênero
df_genero = df_esporte[['Principal esporte praticado', 'Homens', 'Mulheres']].set_index('Principal esporte praticado')

# Definindo as cores
colors = ['blue', 'red']

# Plotando o gráfico
ax = df_genero.plot(kind='bar', color=colors, figsize=(12, 9))
ax.set_title('Quantidade de Esporte Praticado por Gênero')
ax.set_xlabel('Esporte')
ax.set_ylabel('Quantidade')
ax.legend(title='Gênero')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Adicionando os rótulos nas barras
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge')

plt.tight_layout()
plt.show()
```

Considerado o país do futebol masculino, já que o feminino não tem visibilidade no nosso país, não é de se estranhar que a esmagadora maioria dos homens tenha como principal esporte o futebol, escolha de 39% do total dos entrevistados. Entendo que grande parte disso decorre da imensa visibilidade que é dada ao esporte. 
Esse gráfico deixa evidenciado a necessidade de maior visibilidade e divulgação dos outros esportes, que para a população geral, que não acompanha determinado esporte, só é apresentado quando em alguma final de campeonato. Essa falta de visibilidade gera também falta de conhecimeto e interesse nas outras áreas. 

```python
# Gráfico de quantidade por faixa etária
df_idade = df_esporte[['Principal esporte praticado', '15 a 17 anos', '18 a 24 anos', '25 a 39 anos', '40 a 59 anos', '60 anos ou mais']].set_index('Principal esporte praticado')

# Plotando o gráfico
plt.figure(figsize=(12, 8))
df_idade.plot(kind='bar')
plt.title('Quantidade de Esporte Praticado por Faixa Etária')
plt.xlabel('Esporte')
plt.ylabel('Quantidade')
plt.legend(title='Faixa Etária')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Podemos observar que há um certo equilibrio de idade na pratica da maioria dos esportes. Mas na pratica do futebol e da caminhada vemos um diferença considerável, sendo menor no cenário da caminhada. No futebol temos os três grupos mais jovens liderando o rank, enquanto o grupo 60+ aprece timidamente. Já na caminhada quem lidera é o grupo 40 a 59, enquanto o grupo 25 a 39 e 60+ vem logo atrás. O que chama a atenção é a grande presença do grupo 25 a 39 anos em todos os esportes.
O insight que tiro daqui é que é preciso desenvolver mecanismos que incentivem mais a prática dos outros esportes. O futebol e caminhada lideram, enquanto os outros aparecem timidamente. 

```python
# visuzliando o nome das colunas para não haver erro na construção do gráfico
df_final.columns
```

```python

# Filtrar dados da região Nordeste
df_nordeste = df_final[[' Frequência da prática de esporte', 'motivo de terem praticado esporte', 'Nordeste _x', 'Nordeste _y']]

# Agrupar dados por frequência e motivo e somar as quantidades
df_grouped = df_nordeste.groupby([' Frequência da prática de esporte', 'motivo de terem praticado esporte']).mean().reset_index()

# Criar o gráfico de barras empilhadas
plt.figure(figsize=(14, 8))
ax = df_grouped.pivot(index=' Frequência da prática de esporte', columns='motivo de terem praticado esporte', values='Nordeste _x').plot(kind='bar', stacked=True, figsize=(14, 8))

# Personalizar o gráfico
plt.title('Quantidade média de pessoas com frequência da Prática de Esporte e Motivo na Região Nordeste')
plt.xlabel('Frequência da Prática de Esporte')
plt.ylabel('Quantidade')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Motivo')
plt.tight_layout()

# Adicionar rótulos às barras
for container in ax.containers:
    ax.bar_label(container, label_type='center')

# Exibir o gráfico
plt.show()
```

Podemos identificar que pessoas que realizam prática de atividade física por motivos de indicação médica são as que tem a menor frequência, enquanto quem vai para relaxar e se divertir são que possuem maior frequência. Havendo um diferença gritante entre elas. 
Algo que pode ser feito é o incentivo de enxergar a prática esportiva não como uma obrigação, mas como algo prazeroso que gera saúde e bem-estar.
